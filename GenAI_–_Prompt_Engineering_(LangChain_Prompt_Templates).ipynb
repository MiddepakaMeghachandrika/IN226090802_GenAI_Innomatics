Task 4: ChatPromptTemplate System

```python
def get_chat_prompt(role, topic):

    if role == "teacher":
        system_template = "You are a helpful teacher who explains concepts clearly."

    elif role == "interviewer":
        system_template = "You are a strict interviewer who asks challenging questions."

    elif role == "motivator":
        system_template = "You are a motivational speaker who inspires learning."

    else:
        system_template = "You are a helpful assistant."

    system_message = SystemMessagePromptTemplate.from_template(system_template)

    human_message = HumanMessagePromptTemplate.from_template(
        "Talk about {topic}"
    )

    chat_prompt = ChatPromptTemplate.from_messages([system_message, human_message])

    return chat_prompt.format_prompt(topic=topic).to_messages()

# Test
messages = get_chat_prompt("teacher", "Neural Networks")
print(messages)
```

Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience. Defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone. Defaulting to 'formal'")
        tone = "formal"

    return audience, tone

# Test
audience, tone = validate_inputs("kids", "friendly")
print(audience, tone)
```

Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):

    # Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Style-based templates
    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} tone step by step"
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask interview questions about {topic} for {audience} in a {tone} tone"
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )

    else:
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} tone"
        )

    return template.format(topic=topic, audience=audience, tone=tone)

# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
