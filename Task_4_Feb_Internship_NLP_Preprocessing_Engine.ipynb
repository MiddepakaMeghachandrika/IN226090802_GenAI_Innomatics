helper function


```python
def remove_urls_emails(text):
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    return text


def remove_numbers(text):
    return re.sub(r"\d+", "", text)


def normalize_repeated_chars(text):
    # sooo -> soo
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    return text


def remove_extra_spaces(text):
    return re.sub(r'\s+', ' ', text).strip()


def remove_emojis(text):

    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
        "]+",
        flags=re.UNICODE
    )

    return emoji_pattern.sub(r'', text)


def remove_special_characters(text):
    return re.sub(r"[^a-z\s]", "", text)


def remove_short_tokens(tokens):

    important_words = {"no", "not"}

    clean_tokens = []

    for word in tokens:

        if len(word) > 2 or word in important_words:
            clean_tokens.append(word)

    return clean_tokens

```

MAIN PREPROCESS FUNCTION

```python
def preprocess_text(text):

    # error handling
    if not isinstance(text, str):
        return [], ""

    # lowercase
    text = text.lower()

    # remove urls
    text = remove_urls_emails(text)

    # remove numbers
    text = remove_numbers(text)

    # normalize repeated characters
    text = normalize_repeated_chars(text)

    # remove emojis
    text = remove_emojis(text)

    # remove special characters
    text = remove_special_characters(text)

    # remove extra spaces
    text = remove_extra_spaces(text)

    # tokenization
    tokens = text.split()

    # remove small words
    tokens = remove_short_tokens(tokens)

    # final sentence
    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence

```

Task 3: Stress Testing

```python
print("\n\n---- STRESS TESTING ----")

test_sentences = [

    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this",

    "",                 # empty string

    "1234567890"        # only numbers

]

results = []

for sentence in test_sentences:

    tokens, clean_sentence = preprocess_text(sentence)

    token_count = len(tokens)

    unique_tokens = len(set(tokens))

    if token_count > 0:
        avg_length = sum(len(word) for word in tokens) / token_count
    else:
        avg_length = 0

    result = {

        "original": sentence,
        "tokens": tokens,
        "clean_sentence": clean_sentence,
        "token_count": token_count,
        "unique_tokens": unique_tokens,
        "avg_length": avg_length
    }

    results.append(result)


# display output

for r in results:

    print("\n----------------------------------")

    print("Original Text:")
    print(r["original"])

    print("\nClean Tokens:")
    print(r["tokens"])

    print("\nClean Sentence:")
    print(r["clean_sentence"])

    print("\nToken Count:", r["token_count"])

    print("Unique Tokens:", r["unique_tokens"])

    print("Average Token Length:", round(r["avg_length"], 2))

```

 Task 4: Token Analytics

```python
print("\n\n---- TOKEN ANALYTICS ----")

noise_scores = []

for r in results:

    noise_score = len(r["original"]) - len(r["clean_sentence"])

    noise_scores.append(noise_score)


most_noisy_sentence = test_sentences[noise_scores.index(max(noise_scores))]


max_tokens_sentence = max(results, key=lambda x: x["token_count"])


print("\nSentence with most noise:")
print(most_noisy_sentence)


print("\nSentence with most meaningful tokens:")
print(max_tokens_sentence["original"])

```

Task 5: Frequency Analysis

```python
print("\n\n---- FREQUENCY ANALYSIS ----")

all_tokens = []

for r in results:

    all_tokens.extend(r["tokens"])


frequency = Counter(all_tokens)


print("\nTop 10 frequent words:")

for word, count in frequency.most_common(10):

    print(word, ":", count)


print("\nTop 5 least frequent words:")

least_common = frequency.most_common()[::-1][:5]

for word, count in least_common:

    print(word, ":", count)
```

Task 6: Full Pipeline

```python
def full_pipeline(text_list):

    all_tokens = []

    clean_sentences = []

    for text in text_list:

        tokens, clean_sentence = preprocess_text(text)

        all_tokens.extend(tokens)

        clean_sentences.append(clean_sentence)


    return {

        "tokens": all_tokens,

        "clean_sentences": clean_sentences

    }


print("\n\n---- FULL PIPELINE OUTPUT ----")

pipeline_output = full_pipeline(test_sentences)

print(pipeline_output)


```
